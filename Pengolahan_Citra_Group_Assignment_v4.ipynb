# Import dependencies

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import os
from google.colab.patches import cv2_imshow
```

# Load Image

```python
img_original = cv2.imread('original.jpg', cv2.IMREAD_COLOR)

plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.imread('original.jpg', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.title('Grayscale')
plt.axis('off')
plt.show()

cv2.imwrite('grayscale.jpg', img_gray)
```

```python
img_gray = cv2.imread('grayscale.jpg', cv2.IMREAD_GRAYSCALE)

threshold_value = 127

ret, thresholded_img = cv2.threshold(img_gray, threshold_value, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8, 6))
plt.imshow(thresholded_img, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
plt.show()

cv2.imwrite('thresholded_image.jpg', thresholded_img)
```

# Deteksi Tepi

```python
edges = cv2.Canny(thresholded_img, 50, 150, apertureSize=3)

plt.figure(figsize=(8,6))
plt.imshow(edges, cmap='gray')
plt.title('Deteksi Tepi')
plt.axis('off')
plt.show()

cv2.imwrite('canny_edges.jpg', edges)
```

```python
kernel = np.ones((12,12),np.uint8)
closing = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(8,6))
plt.imshow(closing, cmap='gray')
plt.title('Morphological Closing')
plt.axis('off')
plt.show()

cv2.imwrite('closing.jpg', closing)
```

# Compression

```python
closing_image = Image.fromarray(closing)

compressed_image = closing_image.resize((closing_image.width // 2, closing_image.height // 2))

plt.figure(figsize=(8,6))
plt.imshow(np.array(compressed_image), cmap='gray')
plt.title('Compressed Image')
plt.axis('off')
plt.show()

compressed_image.save("compressed_img.jpg")

original_size = os.path.getsize('closing.jpg')
compressed_size = os.path.getsize('compressed_img.jpg')

compression_ratio_kb = (original_size / 1024) / (compressed_size / 1024)

print(f"Original image size: {original_size/1024:.2f} KB")
print(f"Compressed image size: {compressed_size/1024:.2f} KB")
print(f"Compression ratio: {compression_ratio_kb:.2f}")
```
